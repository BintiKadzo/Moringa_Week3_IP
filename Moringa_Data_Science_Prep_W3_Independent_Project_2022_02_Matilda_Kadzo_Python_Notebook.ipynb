**DATA CLEANING**

```python
# Combining data frames df3 - df5
df_combined = pd.concat([df3,df4,df5], axis = 0)
df_combined
```

```python
# getting the full summary of the data frame
df_combined.info()
```

```python
# change the DATE_TIME to datetime
df_combined['DATE_TIME'] = pd.to_datetime(df_combined['DATE_TIME'])
# View the change
df_combined.info()
```

```python
# Drop columns(COUNTRY A and B)
df_combined.drop(['COUNTRY_A', 'COUNTRY_B'], axis = 1)
```

```python
# Drop all duplicates in the dataframe
df_combined = df_combined.drop_duplicates()
```

```python
# View of the aftermath of dropping duplicates
df_combined.info()

# Rows of duplicates dropped = Previous rows - Current rows: 15003-14750 = 253 rows
```

```python
# join the top df1 and df_combined
df_joined = df1.join(df_combined,lsuffix = "_left", rsuffix = "_right")
df_joined
```
